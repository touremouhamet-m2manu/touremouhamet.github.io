```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import LogLocator
```

```python
# Lecture du fichier
df = pd.read_excel('Benzenoides_sym_keku_peri_modif.xlsx')

# Transformation longue
df_melted = df.melt(id_vars='Symétries', 
                    value_vars=['Benzenoides_Kekuléens', 'Benzenoides_non_kekuléens', 'Catacondenses', 'Péricondenses', 'Coronoides'],
                    var_name='Catégorie', 
                    value_name='Total')
```

```python
# Renommer les catégories
df_melted['Catégorie'] = df_melted['Catégorie'].replace({
    'Benzenoides_Kekuléens': 'Kekuleans Benzenoids',
    'Benzenoides_non_kekuléens': 'non-Kekuleans Benzenoids',
    'Catacondenses': 'Catacondensed',
    'Péricondenses': 'Pericondensed',
    'Coronoides': 'Coronoids'
})
```

```python
# Création de la figure
fig, ax = plt.subplots(figsize=(12, 6))

# Palette personnalisée
palette = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

# Paramètres des barres
categories = df_melted['Catégorie'].unique()
symmetries = df_melted['Symétries'].unique()
n_cat = len(categories)
bar_width = 0.2
group_width = bar_width * n_cat
group_spacing = 0.1

# Tracer les barres
for i, cat in enumerate(categories):
    cat_data = df_melted[df_melted['Catégorie'] == cat]
    x = np.arange(len(symmetries)) * (group_width + group_spacing) + i * bar_width - group_width / 2 + bar_width / 2
    ax.bar(x, cat_data['Total'], width=bar_width, label=cat, color=palette[i], edgecolor='black', linewidth=0.4)

# Définir les positions X des ticks (une par symétrie)
xtick_positions = np.arange(len(symmetries)) * (group_width + group_spacing) + group_width / 2 - bar_width / 2

# Appliquer les ticks
ax.set_xticks(xtick_positions)
ax.set_xticklabels(symmetries, fontsize=14)
ax.set_xlabel('Symétries', fontsize=16)
ax.tick_params(axis='x', direction='inout', length=6, width=1, bottom=True, top=False)

# Échelle Y log
ax.set_yscale('log')
ax.set_ylabel('Nombre de benzenoïdes', fontsize=16)
ax.tick_params(axis='y', direction='inout', length=6, width=1, left=True, right=False)

# Lignes verticales fines aux positions des ticks X
for xpos in xtick_positions:
    ax.axvline(x=xpos, color='gray', linestyle=':', linewidth=0.3, alpha=0.7)

# Ticks Y (majeurs et mineurs)
ax.yaxis.set_major_locator(LogLocator(base=10.0, subs=(1.0,), numticks=10))
ax.yaxis.set_minor_locator(LogLocator(base=10.0, subs=np.arange(2, 10)*0.1, numticks=10))
ax.yaxis.set_minor_formatter(plt.NullFormatter())

# Lignes horizontales fines aux ticks majeurs Y
for yval in [1, 10, 100, 1000, 10000]:
    ax.axhline(y=yval, color='gray', linestyle=':', linewidth=0.3, alpha=0.7)

# Limites X
ax.set_xlim(-group_width / 2 - 0.1, len(symmetries) * (group_width + group_spacing) - group_spacing + group_width / 2 + 0.1)

# Légende
ax.legend()

# Affichage et sauvegarde
plt.tight_layout()
#plt.savefig("Figure_symetrie", dpi=300)
plt.show()
```
